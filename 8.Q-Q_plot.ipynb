## Q-Q plot

```python
#Q-Q plot

import numpy as np
import pylab
import scipy.stats as stats

#N(0,1)
std_normal=np.random.normal(loc=0,scale=1,size=1000)    #loc=mean(mu)   scale=std deviation(sigma^2)  size= no. of samples

#0 to 100th percdentile of standard normal
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
import matplotlib.pyplot as plt
```

```python
#generate 100 samples from n(20,5)
measurments=np.random.normal(loc=20,scale=5,size=1000)

# plt.xlim(-1,1)

stats.probplot(measurments,dist="norm",plot=pylab)
pylab.show()
```

```python

measurments=np.random.uniform(low=-1,high=1,size=1000)

# plt.xlim(-4,4)

stats.probplot(measurments,dist="norm",plot=pylab)
pylab.show()
```
